```python
import pandas as pd
food_access_data = pd.read_csv('../data/data-clean/clean_food_access_2019.csv')

# Define the columns for summation and averaging
columns_sum = ['Pop2010', 'OHU2010', 'lapophalf', 'lalowihalf', 'lahunvhalf', 'lapop1', 'lalowi1', 'lahunv1']
columns_avg = ['MedianFamilyIncome']
               #, 'lapophalfshare', 'lalowihalfshare', 'lahunvhalfshare', 'lapop1share', 'lalowi1share', 'lahunv1share']

# Group by 'County' and calculate the required metrics
county_grouped = food_access_data.groupby(['State', 'County']).agg(
    Urban_Percentage=('Urban', lambda x: (x == 1).mean() * 100),
    **{f'{col}_Sum': (col, 'sum') for col in columns_sum},
    **{f'{col}_Avg': (col, 'mean') for col in columns_avg}
).reset_index()

# Save the grouped data to a new CSV file
# county_grouped.to_csv('county_grouped_data.csv', index=False)

county_grouped.head()

```

```python
import pandas as pd
food_access_data = pd.read_csv('clean_food_access_2019.csv')

# Define the columns for summation and averaging
columns_sum = ['Pop2010', 'OHU2010', 'lapophalf', 'lalowihalf', 'lahunvhalf', 'lapop1', 'lalowi1', 'lahunv1']
columns_avg = ['MedianFamilyIncome', 'lapophalfshare', 'lalowihalfshare', 'lahunvhalfshare', 'lapop1share', 'lalowi1share', 'lahunv1share']

# Group by 'County' and calculate the required metrics
county_grouped = food_access_data.groupby('State').agg(
    Urban_Percentage=('Urban', lambda x: (x == 1).mean() * 100),
    **{f'{col}_Sum': (col, 'sum') for col in columns_sum},
    **{f'{col}_Avg': (col, 'mean') for col in columns_avg}
).reset_index()

# Save the grouped data to a new CSV file
county_grouped.to_csv('state_grouped_data.csv', index=False)

county_grouped.head()

```

```python
import pandas as pd

# List of CSV file paths
csv_files = [
    'updated_access_by_state_2019.csv',
    'food_environ_state_data_pivoted.csv',
    'fruit-state-2019-cleaned.csv',
    'veggie-state-2019-cleaned.csv',
    'obesity-state-2019-cleaned.csv'
]

state_column_map = {
    'State': 'State',
    'LocationAbbr': 'State',
}


merged_df = pd.DataFrame()

for csv_file in csv_files:
    # Read the current CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # Identify and rename the state column to a standard name
    for col in df.columns:
        if col in state_column_map:
            df.rename(columns={col: state_column_map[col]}, inplace=True)
            break

    # Merge the current DataFrame with the merged_df on the standardized "states" column
    if merged_df.empty:
        merged_df = df  # For the first file, just assign it
    else:
        merged_df = pd.merge(merged_df, df, on='State', how='outer')

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_dataset.csv', index=False)
```

```python
#county cleaning

import pandas as pd

# List of CSV file paths
csv_files = [
    'food_access_county_2019_updated.csv',
    'food_environ_county_2019.csv'
]

state_column_map = {
    'State': 'State',
    'LocationAbbr': 'State',
}

merged_df = pd.DataFrame()

for csv_file in csv_files:
    # Read the current CSV file into a DataFrame
    df = pd.read_csv("../data/data-county-for-processing/" + csv_file)

    # Identify and rename the state column to a standard name
    for col in df.columns:
        if col in state_column_map:
            df.rename(columns={col: state_column_map[col]}, inplace=True)
            break

    # print(df.head())
    # Define the substrings to remove
    substrings_to_remove = [
        "County", "Borough", "City and Borough", 
        "Municipality", "Census Area"
    ]
    
    # Replace substrings in the 'County' column
    for substring in substrings_to_remove:
        df['County'] = df['County'].str.replace(substring, '', regex=False)
    
    # Optional: Remove any leading or trailing whitespace that might remain after replacement
    df['County'] = df['County'].str.strip()

    # Merge the current DataFrame with the merged_df on the standardized "states" column
    if merged_df.empty:
        merged_df = df  # For the first file, just assign it
    else:
        merged_df = pd.merge(merged_df, df, on=['State', 'County'], how='outer')

merged_df = merged_df.drop("Unnamed: 0")
# Save the merged DataFrame to a new CSV file
# merged_df.to_csv('merged_dataset.csv', index=False)

merged_df[(merged_df['State'] == 'CA')]
```
